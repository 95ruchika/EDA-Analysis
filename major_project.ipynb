```python
#preprocessing
#we clean the data(delete unnecessary columns and rows(NaN))
#unnecessary columns are removed
data = data.drop(['URL'], axis = 1)
data = data.drop(['Offer'], axis = 1)
data
```

```python
#Rows with Column Rating Value - NaN are removed
data.dropna(axis = 0, inplace = True)
data
```

```python
#Data Visualisation-Creation of graphs
```

```python
#For categorical column 'Rating',below it has aggragated the number of times each value occured using the value_counts()
#Below graph designed as X axis Vs.Y Axis - Rating Vs. Number of Restaurants
data['Rating'].value_counts().plot()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,20))
sns.barplot(x=data['Cost for Two (in Rupees)'].value_counts().head(15).values,y=data['Category'].value_counts().head(15).index)
plt.xlabel('Cost for Two (in Rupees)')
plt.ylabel('Category')
#This bar graph represents the top 15 categories of restaurant based on Cost for Two (in Rupees)
```

```python
#divide the data into input and output
#input(x)
#x represents the restaurant name
#output(y)
#y represents the category
#slicing
x=data.iloc[:,0:1].values
x
```

```python
y=data.iloc[:,1].values
y
```

```python
#Train and Test variables
#The train and test split is used to estimate the performance of machine learning algorithms that are applicable for prediction-based algorithms.
#By default the test set is split into 30% of actual data and the training test is split into 70% of actual data
#The reshape() function shapes the dimension of an array without changing the data of an array.
#here we have used reshape(-1,1) which means the resulting array has only one column.
import numpy as np
X=np.array(data["Rating"]).reshape(-1,1)
X
```

```python

y=np.array(data["Cost for Two (in Rupees)"]).reshape(-1,1)
y
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#Run a regressor
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#fit the model(mapping/plotting the inputs with outputs)
model.fit(X_train, y_train)
```

```python
#Predict the output
X_test #actual output values
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
#random_state arguement
#This parameter used to control the shuffling applied to the data before applying the split.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
X_test
```

```python
#Accuracy of the model
#score() it is used to check the accuracy of the model
model.score(X_test, y_test)
```
